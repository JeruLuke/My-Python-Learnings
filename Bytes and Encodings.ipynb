# Bytes and Encodings 

```python
s = 'abcd'
len(s)
```

Each character takes up 1 byte. Languages of Greek, Arabic, Hebrew and so on take up to 2 bytes per character.

Assign a unique number for every character. These codes are encoded in `utf-8` format.

```python
print(s[0])
```

```python
b = b'abc'
type(b)
```

```python
print(b[0])
```

```python
b = bytes(s, encoding='utf-8')
print(b)
```

```python
str(b, encoding = 'utf-8')
```
